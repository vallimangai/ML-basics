```python
import pandas as pd
from sklearn.svm import SVC  
from sklearn.model_selection import train_test_split
from sklearn import metrics
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('iris.csv')
```

```python
df
```

```python
X=df.drop('species',axis=1)
y=df[['species']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/5, random_state = 47)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid= {'decision_function_shape':['ovr','ovo'],
              'kernel': ['linear','rbf','poly','sigmoid']}
```

```python
grid = GridSearchCV(SVC(), param_grid) 
grid.fit(X_train, y_train)
print(grid.best_params_)
```

```python
print(grid.best_params_)
```

```python
df = pd.DataFrame(grid.cv_results_)
df
```

```python
df[['param_decision_function_shape','param_kernel','mean_test_score']]
```

```python
decision_function_shape=['ovr','ovo']
kernel=['linear','rbf','poly','sigmoid']

for kval in kernel:
    for cval in decision_function_shape:
        clf = SVC(decision_function_shape =cval,kernel=kval) 
        clf.fit(X_train, y_train)
        y_pred = clf.predict(X_test)
        print('#'*50)
        print(cval," ",kval)
        confusion = metrics.confusion_matrix(y_test, y_pred)
        print(confusion)
        print('Accuracy Score on confusion_matrix: ' ,metrics.accuracy_score(y_test, y_pred))
        print(metrics.classification_report(y_test, y_pred))

```
