## Questions:
#### Implement Support Vector Machine Algorithm for Multi-class classification with the choice of having different kernels.

```python
import pandas as pd
from sklearn.svm import SVC  
from sklearn.model_selection import train_test_split
from sklearn import metrics
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_openml
mnist=fetch_openml('mnist_784',version=1)
```

```python
mnist.keys()
```

```python
X,y=mnist["data"][:10000],mnist["target"][:10000]
print(X.shape)
print(y.shape)
```

```python
X.info()
```

```python
X.head()
```

### Description
This data set is conside of handwritten numbers . (28px X 28px)

our target is to find the number written

```python
y[0]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/5, random_state = 47)
```

```python
clf = SVC() 
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
```

```python
confusion = metrics.confusion_matrix(y_test, y_pred)
print(confusion)
print('Accuracy Score on confusion_matrix: ' ,metrics.accuracy_score(y_test, y_pred))
print(metrics.classification_report(y_test, y_pred))
```

```python
clf = SVC(kernel='linear')
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
confusion = metrics.confusion_matrix(y_test, y_pred)
print(confusion)
print('Accuracy Score on confusion_matrix: ' ,metrics.accuracy_score(y_test, y_pred))
print(metrics.classification_report(y_test, y_pred))
```

```python
clf = SVC(kernel='rbf')
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
confusion = metrics.confusion_matrix(y_test, y_pred)
print(confusion)
print('Accuracy Score on confusion_matrix: ' ,metrics.accuracy_score(y_test, y_pred))
print(metrics.classification_report(y_test, y_pred))
```

```python
clf=SVC(kernel='sigmoid')
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
confusion = metrics.confusion_matrix(y_test, y_pred)
print(confusion)
print('Accuracy Score on confusion_matrix: ' ,metrics.accuracy_score(y_test, y_pred))
print(metrics.classification_report(y_test, y_pred))
```

```python
clf=SVC(kernel='poly', degree=8)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
confusion = metrics.confusion_matrix(y_test, y_pred)
print(confusion)
print('Accuracy Score on confusion_matrix: ' ,metrics.accuracy_score(y_test, y_pred))
print(metrics.classification_report(y_test, y_pred))
```
